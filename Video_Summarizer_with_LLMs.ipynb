```python
from youtube_transcript_api import YouTubeTranscriptApi
```

```python
def get_video_id(url_link):
  return url_link.split("watch?v=")[-1]
```

```python
video_id = get_video_id("https://www.youtube.com/watch?v=eMlx5fFNoYc")
```

```python
transcript = YouTubeTranscriptApi.get_transcript(video_id)
```

```python
transcript
```

```python
transcript_joined = " ".join([line['text'] for line in transcript])
```

```python
prompt = f'"Summarize this text: \ntext = "{transcript_joined}"'
```

```python
from openai import OpenAI

client = OpenAI(
    api_key=api_key # This is the default and can be ommited

)

chat_completion = client.chat.completions.create(
    messages=[
        {
            "role": "user",
            "content": prompt,
        }
    ],
    model="gpt-3.5-turbo",
    temperature=1, # Temperature controls randomness in the response
    max_tokens=128, # Maximum number of tokens in the response
    top_p=1, #Top-p (nucleus) samping parameter, higher values make output more focussed
    frequency_penalty=0, # Frequency penalty discourages the model from repeating words or phrases
    presence_penalty=0 # Presence penalty discourages the model from adding verbose or necessary words
)
```

```python
import google.generativeai as genai

# Set up your Gemini API key
api_key = ""

# Configure the client
genai.configure(api_key=api_key)

# Define the prompt
prompt = f"Summarize this text:\n{transcript_joined}"

# Use the correct Gemini model
model = genai.GenerativeModel(model_name="gemini-2.0-flash")  # Ensure correct model name

# Generate the response
response = model.generate_content(prompt)

# Print the output
print(response.text)
```
